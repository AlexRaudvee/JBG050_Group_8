```python
import pandas as pd
from datetime import datetime

# Define the path to the CSV file
csv_file_path = r'data/PAS_data_ward_level/PAS_ward_level_FY_20_21.csv'

# Define the column name for borough
borough_column = 'Borough' if csv_file_path[-9:-4] == '20_21' else 'BOROUGHNEIGHBOURHOODCODED'

# Read the CSV file
df = pd.read_csv(csv_file_path)

# Function to extract the month and year
def convert_date_format(date_str):
    # Split the string by parentheses and extract the part containing the month and year
    month_year_str = date_str.split('(')[1].split(')')[0]
    # Parse the month and year and format it as 'YYYY-MM'
    return datetime.strptime(month_year_str, '%b %Y').strftime('%Y-%m')

# Apply the function to the MONTH column
df['MONTH'] = df['MONTH'].apply(convert_date_format)

# Group the DataFrame by the "Borough" column
grouped_df = df.groupby([borough_column, "MONTH"])

# Create a new DataFrame to store the results
results_df = pd.DataFrame(columns=['Date', 'Borough', 'Measure', 'Total Proportion', 'White_British_Proportion', 'White_Other_Proportion', 'Black_Proportion', 'Asian_Proportion', 'Mixed_Proportion', 'Other_Proportion'])

# Iterate over each question and borough to calculate proportions
for question in questions_dict.keys():
    for (borough, month), group_df in grouped_df:
        date = month
        measure = question
        borough = borough
        try:
            # Calculate the total proportion
            if question in weighted_questions:
                value_counts = group_df[question].value_counts(normalize=True)
                total_proportion = sum(value_counts[category] * weights.get(category, 0) for category in value_counts.index)
            else:
                value_counts = group_df[question].value_counts(normalize=True)
                total_proportion = sum(value_counts[category] * weights.get(category, 0) for category in value_counts.index)
            
            # Extract the ethnic group proportions
            white_british_proportion = group_df[group_df['ReNQ147'] == 'White British'].shape[0] / group_df.shape[0]
            black_proportion = group_df[group_df['ReNQ147'] == 'Black'].shape[0] / group_df.shape[0]
            asian_proportion = group_df[group_df['ReNQ147'] == 'Asian'].shape[0] / group_df.shape[0]
            white_other_proportion = group_df[group_df['ReNQ147'] == 'White Other'].shape[0] / group_df.shape[0]
            mixed_proportion = group_df[group_df['ReNQ147'] == 'Mixed'].shape[0] / group_df.shape[0]
            other_ethnicity_proportion = 1 - white_british_proportion - white_other_proportion - black_proportion - asian_proportion  # Assuming there are only 'White' and 'Black' ethnicities
            
            # Append the results to the DataFrame
            results_df = results_df.append({
                'Date': date,
                'Borough': borough,
                'Measure': measure,
                'Total Proportion': total_proportion,
                'White_British_Proportion': white_british_proportion,
                'White_Other_Proportion': white_other_proportion,
                'Black_Proportion': black_proportion,
                'Asian_Proportion': asian_proportion,
                'Mixed_Proportion': mixed_proportion,
                'Other_Proportion': other_ethnicity_proportion
            }, ignore_index=True)
        except KeyError as e:
            print(f"given question ({question}) didn't occur in {csv_file_path[-9:-4]}")

# Print the resulting DataFrame
results_df.to_csv('data/pas_data_ward_level/pre_final.csv')
```

# This code is to find the percentages of the differnet ethnic groups that have bad results for the relevant question in every borough

```python
import pandas as pd
import os

#Define the path to the CSV file
csv_file_path = r'data/pas_data_ward_level/PAS_ward_level_FY_20_21.csv'

#Name of the columns we want to extract values from
column_name = 'Q13'
disagree_values = ['Fairly worried', 'Very worried']  # Specify the values indicating disagreement
if csv_file_path[-9:-4] == '20_21':
    borough_column = 'Borough' 
else:
    borough_column = 'BOROUGHNEIGHBOURHOODCODED'
ethnic_column = 'ReNQ147'  #The column for the ethnic groups

#Read the CSV file
df = pd.read_csv(csv_file_path)

#Group the DataFrame by the "Borough" column
grouped_df = df.groupby(borough_column)

#Initialize a dictionary to store the percentages of ethnic groups for each borough
ethnic_percentages_by_borough = {}

#Iterate over each borough group
for borough, group_df in grouped_df:
    #Filter the DataFrame to include only responses where respondents disagreed or strongly disagreed
    disagreed_df = group_df[group_df[column_name].isin(disagree_values)]
    
    #Count the occurrences of each ethnic group in the filtered DataFrame
    ethnic_counts = disagreed_df[ethnic_column].value_counts()
    
    #Calculate the total number of responses where respondents disagreed or strongly disagreed
    total_responses = ethnic_counts.sum()
    
    #Calculate the percentages of each ethnic group
    ethnic_percentages = (ethnic_counts / total_responses) * 100
    
    #Store the percentages for the current borough
    ethnic_percentages_by_borough[borough] = ethnic_percentages

#Print the percentages by borough
for borough, ethnic_percentages in ethnic_percentages_by_borough.items():
    print(f"\nBorough: {borough}")
    print(ethnic_percentages)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#Define file paths
file_paths = [
    r'data/pas_data_ward_level/PAS_ward_level_FY_20_21.csv',
    r'data/pas_data_ward_level/PAS_ward_level_FY_19_20.csv',
    r'data/pas_data_ward_level/PAS_ward_level_FY_18_19.csv',
    r'data/pas_data_ward_level/PAS_ward_level_FY_17_18.csv'
]

#Define columns and values
column_name = 'Q60'
value1 = 'Poor'
value2 = 'Very poor'

#Function to preprocess borough names
def preprocess_borough(borough_name):
    if isinstance(borough_name, str):  
        return borough_name.replace(' and ', ' & ')
    return borough_name  

#Function to process each file and calculate percentages
def process_file(file_path, borough_col, split_borough=False):
    df = pd.read_csv(file_path)
    if split_borough:
        df[borough_col] = df[borough_col].str.split(' - ').str[0]

    #Preprocess borough names
    df[borough_col] = df[borough_col].apply(preprocess_borough)

    #Filter for specified values
    filtered_df1 = df[df[column_name] == value1]
    filtered_df2 = df[df[column_name] == value2]

    #Count occurrences and calculate percentages
    borough_counts1 = filtered_df1[borough_col].value_counts()
    borough_counts2 = filtered_df2[borough_col].value_counts()
    combined_counts = borough_counts1.add(borough_counts2, fill_value=0)
    total_counts = df[borough_col].value_counts()
    percentages = (combined_counts / total_counts) * 100

    return percentages.sort_values(ascending=False)

#Process each file
percentages_list = []
percentages_list.append(process_file(file_paths[0], 'Borough', split_borough=False))

for file_path in file_paths[1:]:
    percentages_list.append(process_file(file_path, 'BOROUGHNEIGHBOURHOODCODED', split_borough=True))

#Reverse the order of percentages_list to match the desired order in the plots
percentages_list = percentages_list[::-1]

#Find the top 5 boroughs for each file
top_boroughs = [percentages.index[:5] for percentages in percentages_list]

#Create dataframes for the top 5 boroughs and their percentages
top_boroughs_dfs = []
for i, percentages in enumerate(percentages_list):
    top_boroughs_df = pd.DataFrame({
        'Borough': percentages.index[:5],
        'Percentage': percentages.values[:5]
    })
    top_boroughs_dfs.append(top_boroughs_df)
    print(f"Top 5 Boroughs for financial year {17+i} - {18+i}:\n", top_boroughs_df, "\n")

#Create line plots for the top 5 boroughs in each file
for i, top_boroughs_list in enumerate(top_boroughs):
    plt.figure(figsize=(12, 6))
    for borough in top_boroughs_list:
        borough_percentages = []
        for percentages in percentages_list:
            borough_percentages.append(percentages.get(borough, 0))
        plt.plot(range(1, 5), borough_percentages, marker='o', label=borough)
    
    plt.title(f'Top 5 Boroughs for financial year {17+i} - {18+i} Across All Years')
    plt.xlabel('Financial Year')
    plt.ylabel('Percentage')
    plt.xticks(range(1, 5), ['17-18','18-19','19-20','20-21'])  # Adjusting x-axis labels to show the correct order
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

# This code creates a dataframe for the best/ worst 3 boroughs for each ethnic group

```python
import pandas as pd

#Define the path to the CSV file
csv_file_path = r'data/pas_data_ward_level/PAS_ward_level_FY_20_21.csv'

#Name of the columns we want to extract values from
column_name = 'NQ135BD'
disagree_values = ['Strongly disagree', 'Tend to disagree']  #Specify the values indicating disagreement
borough_column = 'Borough'
ethnic_column = 'ReNQ147'  # the column for the ethnic groups

#Read the CSV file
df = pd.read_csv(csv_file_path)

#Initialize a dictionary to store the percentages of ethnic groups for each borough
ethnic_percentages_by_borough = {}

#Group the DataFrame by the "Borough" column
grouped_df = df.groupby(borough_column)

#Iterate over each borough group
for borough, group_df in grouped_df:
    #Filter the DataFrame to include only responses where respondents disagreed or strongly disagreed
    disagreed_df = group_df[group_df[column_name].isin(disagree_values)]
    
    #Count the occurrences of each ethnic group in the filtered DataFrame
    ethnic_counts = disagreed_df[ethnic_column].value_counts()
    
    #Calculate the total number of responses where respondents disagreed or strongly disagreed
    total_responses = ethnic_counts.sum()
    
    #Calculate the percentages of each ethnic group
    ethnic_percentages = (ethnic_counts / total_responses) * 100
    
    #Store the percentages for the current borough
    ethnic_percentages_by_borough[borough] = ethnic_percentages

#Convert the dictionary to a DataFrame
ethnic_percentages_df = pd.DataFrame(ethnic_percentages_by_borough).transpose().fillna(0)

#Initialize dictionaries to store top and bottom 3 boroughs with percentages
top_bottom_boroughs = {
    'Ethnic Group': [],
    'Top 3 Boroughs': [],
    'Top 3 Percentages': [],
    'Bottom 3 Boroughs': [],
    'Bottom 3 Percentages': []
}

#Find the top 3 and bottom 3 boroughs for each ethnic group
for ethnic_group in ethnic_percentages_df.columns:
    sorted_boroughs = ethnic_percentages_df[ethnic_group].sort_values(ascending=False)
    top_3 = sorted_boroughs.head(3).index
    bottom_3 = sorted_boroughs.tail(3).index
    top_3_percentages = sorted_boroughs.head(3).values
    bottom_3_percentages = sorted_boroughs.tail(3).values
    
    top_bottom_boroughs['Ethnic Group'].append(ethnic_group)
    top_bottom_boroughs['Top 3 Boroughs'].append(', '.join(top_3))
    top_bottom_boroughs['Top 3 Percentages'].append(', '.join(f"{percent:.2f}%" for percent in top_3_percentages))
    top_bottom_boroughs['Bottom 3 Boroughs'].append(', '.join(bottom_3))
    top_bottom_boroughs['Bottom 3 Percentages'].append(', '.join(f"{percent:.2f}%" for percent in bottom_3_percentages))

#Create a DataFrame to display the results
top_bottom_boroughs_df = pd.DataFrame(top_bottom_boroughs)
top_bottom_boroughs_df
```

# This code gets you the highest given answer + its percentage in every borough for a specific question

```python
import pandas as pd

#Define the path to the CSV file
csv_file_path = r'data/pas_data_ward_level/PAS_ward_level_FY_20_21.csv'

#Name of the columns we want to extract values from
column_name = 'XQ122B'
borough_column = 'Borough'

#Read the CSV file
df = pd.read_csv(csv_file_path)

#Group the DataFrame by the "Borough" column
grouped_df = df.groupby(borough_column)

#Initialize a list to store the result
result_data = []

#Iterate over each borough group
for borough, group_df in grouped_df:
    #Get the occurrences of values for the specified column
    value_counts = group_df[column_name].value_counts()
    
    #Find the highest value and its count
    most_common_answer = value_counts.idxmax()
    most_common_answer_count = value_counts.max()
    
    #Calculate the percentage of the most occurring answer among all the given answers
    total_answers = value_counts.sum()
    percentage = (most_common_answer_count / total_answers) * 100
    result_data.append({'Borough': borough,
                        'Most Common Answer': f'{most_common_answer} ({most_common_answer_count})',
                        'Percentage': percentage})


result_df = pd.DataFrame(result_data)
result_df
```
